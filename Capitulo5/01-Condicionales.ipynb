# Condicionales y recursividad

```python
minutos = 105
horas = minutos // 60
minutos_restantes = minutos % 60
print(f"{minutos} minutos son {horas} horas y {minutos_restantes} minutos.")


print("Expresión booleana")
print(" 5 == 5     -> ", 5 == 5)
print(" 5 != 5     -> ", 5 != 5)
print(" 5 > 3      -> ", 5 > 3)
print(" 5 < 3      -> ", 5 < 3)
print(" 5 >= 5     -> ", 5 >= 5)

print(" type(True) -> ", type(True))

print("Operadores lógicos")
x = 5
print(" x = 5")
print(" x > 0 and x < 10 - > ", x > 0 and x < 10)


x = 6
print(" x = 6")
print("n %2 == 0 or n %3 == 0 -> ", x %2 == 0 or x %3 == 0)


print(" 42 and True -> ", 42 and True)
print(" 0 and True  -> ", 0 and True)

print("Ejecución condicional")

print(" if x > 0:")
print("    print('x es positivo')")

if x > 0:
    print('x es positivo')


print("Ejecución alternativa")
print("if x % 2 == 0:")
print("   print('x es par')")
print("else:")
print("   print('x es impar')")
if x % 2 == 0:
    print('x es par')
else:
    print('x es impar')

y = 4
print("Condicionales encadenados")
if x < y:
    print('x es menor que y')
elif x > y:
    print('x es mayor que y')
else:
    print('x e y son iguales')

opcion = 'c'
if opcion == 'a':
    print("dibujar_a()")
elif opcion == 'b':
    print("ibujar_b()")
elif opcion == 'c':
    print("dibujar_c()")


print("Condicionales anidados")

if x == y:
    print('x e y son iguales')
else:
    if x < y:
        print('x es menor que y')
    else:
        print('x es mayor que y')


if 0 < x and x < 10:
    print('x es un número positivo de un dígito.')

if 0 < x < 10:
    print('x es un número positivo de un dígito.')
```
